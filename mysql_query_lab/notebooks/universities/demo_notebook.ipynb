```js
+--------------+          +-----------------+          +-----------------------+
| id (PK)      |          | id (PK)         |          | id (PK)               |
| country_name |----------| country_id (FK) |          | system_name           |
+--------------+          | university_name |           +-----------------------+
                           +----------------+

+-----------------------+        +---------------------------+
|     ranking_criteria  |        |       university_year     |
+-----------------------+        +---------------------------+
| id (PK)               |        | university_id (FK)        |
| ranking_system_id (FK)|--------| year                      |
| criteria_name         |        | num_students              |
+-----------------------+        | student_staff_ratio       |
                                 | pct_international_students|
                                 | pct_female_students       |
                                 +---------------------------+

+---------------------------+
|  university_ranking_year  |
+---------------------------+
| university_id (FK)        |
| ranking_criteria_id (FK)  |
| year                      |
| score                     |
+---------------------------+

```


````python
%%sql 

SELECT
u.university_name,
c.country_name,
rc.criteria_name,
ury.year,
ury.score
FROM university u
INNER JOIN country c ON u.country_id = c.id
INNER JOIN university_ranking_year ury ON ury.university_id = u.id
INNER JOIN ranking_criteria rc ON ury.ranking_criteria_id = rc.id
WHERE rc.criteria_name IN (
'Total Times',
'Total Shanghai',
'Total CWUR'
)
AND ury.year = 2010
ORDER BY ury.score DESC;
````
